## Exercise 1 – How long are movie titles?

In this exercise, you will apply descriptive statistics techniques to a real dataset: a database of 1000 popular movies from the IMDb website.

From the title of each movie, you will:
- Calculate the title length (in characters).
- Obtain key statistical metrics: mean, median, mode, range, variance, standard deviation, skewness, and kurtosis.
- Identify the longest and shortest titles.
- Visualize the distribution of these lengths with a histogram and a density curve.


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Load IMDb dataset
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
#gives us the title length, but does not add to the visual worksheet
df['title_length'] = 0 

for title in df['title']: 
    df['title_length'] = df['title_length'] 
    print(len(title))
```

```python
# Create a new column called 'title_length' with the length (number of characters) of each title

# 1st example (Pandas code, because we used Dataframes)
df["title_length"] = df["title"].str.len()
df.head(5)
```

```python
#2nd example (for loop)
title_lengths = []

for title in df['title']:
    title_lengths.append(len(title))

df['title_length'] = title_lengths
print(df.head(10))
```

```python
# Calculate mean, median, and mode of the 'title_length' column

mean_length = df['title_length'].mean()
print(f"mean length is {mean_length}")

# mid = np.median(df['title_length'])
mid = df['title_length'].sort_values().reset_index(drop=True)[len(df)//2]
print(f"median length is {mid}")


mode_length = df['title_length'].mode()[0]
print(f"mode length is {mode_length}")

```

```python
# Calculate the range, variance, and standard deviation of 'title_length'
range = df['title_length'].max() - df['title_length'].min()
print(f"range is {range}")

variance = df["title_length"].var()
print(f"variance is {variance}")

standard_d = df["title_length"].std()
print(f"standard_d is {standard_d}")
```

```python
#This is ChatGPT, gave it my chart and ask for a heat map version
import matplotlib.pyplot as plt
import numpy as np

# Sort the data so colors go smoothly from shortest → longest
df_sorted = df.sort_values('title_length')

# Normalize title lengths for coloring
norm = plt.Normalize(df_sorted['title_length'].min(), df_sorted['title_length'].max())
colors = plt.cm.RdYlGn(norm(df_sorted['title_length']))

# Create figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Plot bars
bars = ax.bar(df_sorted['title'], df_sorted['title_length'], color=colors)

# Rotate x-axis labels
ax.set_xticklabels(df_sorted['title'], rotation=90)

# Labels and title
ax.set_xlabel('Movie Title')
ax.set_ylabel('Title Length (characters)')
ax.set_title('Movie Title Lengths (Red = Short, Green = Long)')

# Add colorbar correctly
sm = plt.cm.ScalarMappable(cmap='RdYlGn', norm=norm)
sm.set_array(df_sorted['title_length'])  # this fixes the ValueError
fig.colorbar(sm, ax=ax, label='Title Length')

plt.show()
```
